## Task-1

```
# reference https://towardsdatascience.com/building-a-convolutional-neural-network-cnn-in-keras-329fbbadc5f5
# reference https://elitedatascience.com/keras-tutorial-deep-learning-in-python

from keras.datasets import mnist
import matplotlib.pyplot as plt
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D, Dropout, Input
import numpy as np


def gen_image(arr):
    two_d = (np.reshape(arr, (28, 28)) * 255).astype(np.uint8)
    plt.imshow(two_d, interpolation='nearest')
    return plt



from google.colab import drive
drive.mount('/content/drive')


#download mnist data and split into train and test sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()

#check image shape
print(X_train[0].shape)

#normalization values between 0 and 1
#X_train = X_train / 255
#X_test = X_test / 255

#reshape data to fit model
X_train = X_train.reshape(60000,28,28,1)
X_test = X_test.reshape(10000,28,28,1)

#one-hot encode target column which is equal to generate_t in program 5
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
y_train[0]


#create model
#Add your own CNN model here
# Convulutional layer
model = Sequential()
model.add(Input(shape=(28, 28, 1)))
model.add(Conv2D(2, kernel_size=3, activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Dropout layer
model.add(Dropout(0.25))

# Flatten layer
model.add(Flatten())

# Dense layer
model.add(Dense(128))

# Additional Dropout
model.add(Dropout(0.5))

# Softmax Output layer
model.add(Dense(10, activation='softmax'))


###################

# 8. Compile model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# 9. Fit model on training data
model.fit(X_train, y_train,
          batch_size=32, epochs=10, verbose=1) #epochs  = iterations(Nit)

# 10. Evaluate model on test data
score = model.evaluate(X_test, y_test, verbose=1)

print('Testing accuracy - > ',score[1] * 100)

#ytested = model.predict_classes(X_test)
predicted=model.predict(X_test)


```

```
y_tested=np.argmax(predicted,axis=-1)

for i in range(4):
  gen_image(X_test[i]).show() # printing image vs the predicted image below
  print("The Predicted Testing image is =%s" % (y_tested[i]))
```

## Task-2

```
# Task 2 Model Structure
# Initialize a Sequential model
model2 = Sequential()

# Convolutional Layer 1
model2.add(Conv2D(64, kernel_size=3, activation='relu', input_shape=(28,28,1)))

# Max Pooling Layer (uncommented as per Task 2)
model2.add(MaxPooling2D(pool_size=(2, 2)))

# Dropout Layer
model2.add(Dropout(0.25))

# Flatten Layer
model2.add(Flatten())

# Fully Connected Hidden Layer
model2.add(Dense(128, activation='relu'))

# Additional Dropout Layer
model2.add(Dropout(0.5))

# Softmax Output Layer
model2.add(Dense(10, activation='softmax'))

# Compile the model
model2.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model for 10 epochs
model2.fit(X_train, y_train, batch_size=32, epochs=10, verbose=1)

# Evaluate the model on test data
score = model2.evaluate(X_test, y_test, verbose=1)
print('Testing accuracy ->', score[1] * 100)

# Predictions for the test data
predicted = model2.predict(X_test)
y_tested = np.argmax(predicted, axis=-1)

# Display the first 4 test images and their predicted classes
for i in range(4):
    gen_image(X_test[i]).show()
    print("The Predicted Testing image is =", y_tested[i])
```
